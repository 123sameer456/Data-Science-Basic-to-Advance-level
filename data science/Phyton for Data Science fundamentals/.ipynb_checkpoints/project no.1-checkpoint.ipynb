```python
update_value = {}
for app in android[:]:
    name = app[0]
    n_review = app[3]
    if name in update_value and update_value[name]< n_review:
        update_value[name] = n_review
    elif name not in update_value:
        update_value[name] = n_review
print(update_value["Instagram"])
print (update_value)["Facebook"]
```

```python
android_clean = []
name_appended = []
for a in android:
    name = a[0]
    n_review = a[3]
    if update_value[name]== n_review and name not in name_appended:
        android_clean.append(a)
        name_appended.append(name)
len(android_clean)
```

```python
for app in android_clean:
    print(app[0])
```

```python
def english(name):
    asc = 0
    for i in name:
        if ord(i)>127:
            asc+=1
    if asc >3:
        return False
    else:
        return True
    
```
