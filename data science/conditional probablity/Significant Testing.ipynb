Null_Hypothesis = No change after
ALternate Hypothesis = Change

```python
group_a=[1,2,3,3,3,3,5,1,10,3,7,6,5,4,2] # Control group # candy
group_b=[3,2,6,5,7,4,4,7,7,2,9,10,4,5,2] # treatement group # actual pill
```

Null_Hypothesis = No change in weight
ALternate Hypothesis = lose weight

```python
import numpy as np
import pandas as pd

mean_a=np.mean(group_a)
mean_b=np.mean(group_b)

hypothesis=mean_b-mean_a
hypothesis
```

```python
total_value=[1,2,3,3,3,3,5,1,10,3,7,6,5,4,2,2,2,2,2,2,3,2,7,5,5,5,6,7,3,2,2,10,4,5,2,5,5,5,5,5]
```

```python
len(total_value)
```

```python
differences=[]
for a in range(1000):
    grp_a=[]
    grp_b=[]
    for value in total_value:
        chances=np.random.rand()
        if chances<0.5:
            grp_a.append(value)
        elif chances>0.5:
            grp_b.append(value)
    difference=np.mean(grp_b)-np.mean(grp_a)
    differences.append(difference)
```

```python
differences
```

```python
counter=0
for i in differences:
    if i>1.25:
        counter+=1
        
```
