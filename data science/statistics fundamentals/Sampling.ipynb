```python
wnba_under12 = wnba[wnba["Games Played"] <= 12]
wnba_underbtw13_22 = wnba[(wnba["Games Played"] > 12) & (wnba["Games Played"]<=22)]
wnba_over = wnba[wnba["Games Played"]> 22]
```

```python
sample_mean = []

for i in range(0,100):
    first = wnba_under12["PTS"].sample(1, random_state = i)
    second = wnba_underbtw13_22["PTS"].sample(1, random_state = i)
    third = wnba_over["PTS"].sample(1, random_state = i)
    final_sample = pd.concat([first,second,third], axis = 0)
    statistic = final_sample.mean()
    sample_mean.append(statistic)
```

```python
plt.scatter(range(0,100),sample_mean)
plt.axhline(parameter)
```
