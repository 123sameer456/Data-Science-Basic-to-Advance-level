```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degree = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
print(women_degree.info())
print(women_degree.head())
```

```python
maj_cats = ["Biology","Computer Science","Engineering", "Math and Statistics"]


fig = plt.figure(figsize = (12,12))

for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degree["Year"],women_degree[maj_cats[i]], label = "Women", c = "green")
    ax.plot(women_degree["Year"],100-women_degree[maj_cats[i]], label = "Men", c = "red")
    ax.tick_params(left = "off", bottom = "off", right = "off")
    ax.set_xlim(1970,2011)
    ax.set_ylim(0,100)
    ax.set_title(maj_cats[i])
    #plt.legend()
    for key, spine in ax.spines.items():
        spine.set_visible(False)
ax.legend(loc = "upper right")
```

```python
maj_cats = ["Biology","Computer Science","Engineering", "Math and Statistics"]
dark_blue = (0/255,107/255,164/255) #RGB
orange = (255/255,128/255,14/255)

fig = plt.figure(figsize = (16,3))

for i in range(4):
    ax = fig.add_subplot(1,4,i+1)
    ax.plot(women_degree["Year"],women_degree[maj_cats[i]], label = "Women", c =dark_blue, linewidth = 3)
    ax.plot(women_degree["Year"],100-women_degree[maj_cats[i]], label = "Men", c = orange, linewidth = 3)
    ax.tick_params(left = "off", bottom = "off", right = "off")
    ax.set_xlim(1970,2011)
    ax.set_ylim(0,100)
    ax.set_title(maj_cats[i])
    if i == 0:
        ax.text(2005,66,"Women")
        ax.text(2005,30,"Men")
    if i == 3:
        ax.text(2005,60,"Women")
        ax.text(2005,35,"Men")
    #plt.legend()
    for key, spine in ax.spines.items():
        spine.set_visible(False)
#ax.legend(loc = "upper right")
```
