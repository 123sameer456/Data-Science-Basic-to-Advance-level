```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degree = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
print(women_degree.info())
print(women_degree.head())
```

```python
plt.plot(women_degree["Year"],women_degree["Agriculture"], label = "Women", c = "green")
plt.plot(women_degree["Year"],100-women_degree["Agriculture"], label = "Men", c = "red")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(women_degree["Year"],women_degree["Biology"], label = "Women", c = "green")
plt.plot(women_degree["Year"],100-women_degree["Biology"], label = "Men", c = "red")
ax.set_xlabel("year")
plt.ylabel("percentage")
ax.tick_params(left = "off", bottom = "off")
plt.legend()
plt.title("percentage of biology")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(women_degree["Year"],women_degree["Biology"], label = "Women", c = "green")
plt.plot(women_degree["Year"],100-women_degree["Biology"], label = "Men", c = "red")
ax.set_xlabel("year")
plt.ylabel("percentage")
ax.tick_params(left = "off", bottom = "off")#remove parameter
for key, spine in ax.spines.items():
    spine.set_visible(False)
#ax.spines["right"].set_visible("False")
plt.legend()
plt.title("percentage of biology")
plt.show()
```

```python
maj_cats = ["Biology","Computer Science","Engineering", "Math and Statistics"]


fig = plt.figure(figsize = (12,12))

for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degree["Year"],women_degree[maj_cats[i]], label = "Women", c = "green")
    ax.plot(women_degree["Year"],100-women_degree[maj_cats[i]], label = "Men", c = "red")
    ax.tick_params(left = "off", bottom = "off", right = "off")
    ax.set_xlim(1970,2011)
    ax.set_ylim(0,100)
    ax.set_title(maj_cats[i])
    #plt.legend()
    for key, spine in ax.spines.items():
        spine.set_visible(False)
ax.legend(loc = "upper right")
```
