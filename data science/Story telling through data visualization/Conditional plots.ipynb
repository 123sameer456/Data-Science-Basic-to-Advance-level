```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("train.csv")
print(titanic.info())
print(titanic.head())
```

```python
cols = ["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
```

```python
titanic = titanic[cols]
titanic.info()
```

```python
titanic.dropna(inplace = True)
titanic.info()
```

```python
from mpl_toolkits.basemap import Basemap as bmap
```

```python
import seaborn as sns
sns.distplot(titanic["Fare"])
```

```python

import seaborn as sns
sns.distplot(titanic["Age"])
print(titanic["Age"].min())
print(titanic["Age"].max())
```

```python
sns.kdeplot(titanic["Age"], shade = True)
sns.set_style("ticks")
sns.despine()
```

```python
sns.kdeplot(titanic["Fare"], shade = True)
sns.set_style("ticks")
plt.xlabel("Age")
sns.despine(bottom = True, left = True)
```

#CONDITIONAL PLOTS

```python
g = sns.FacetGrid(titanic, col = "Survived" , size = 5)
g.map(sns.kdeplot, "Age", shade = True)
```

```python

g = sns.FacetGrid(titanic, col = "Pclass" , size = 5)
g.map(sns.kdeplot, "Age", shade = True)
```

```python
g = sns.FacetGrid(titanic, col = "Pclass" , row = "Survived" ,size = 3)
g.map(sns.kdeplot, "Age", shade = True)
```

```python
g = sns.FacetGrid(titanic, col = "Pclass" , row = "Survived", hue = "Sex" ,size = 3)
g.map(sns.kdeplot, "Age", shade = True)
g.add_legend()
sns.despine(bottom = True , left = True)
```
