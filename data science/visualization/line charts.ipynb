```python
import numpy as np
import pandas as pd
unrate = pd.read_csv("unrate.csv")
print(unrate.info())
print(unrate.head())
```

```python
import datetime as dt
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
print(unrate["DATE"].dtype) 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12))
plt.show()
```

```python
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12))
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12))
plt.xticks(rotation = 90)
plt.hlines(3.9,"1948-01","1948-12", color = "red")
plt.show()
```

```python
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12))
plt.xticks(rotation = 90)
plt.hlines(3.5,"1948-01","1948-12", color = "red")
plt.show()
```

```python
plt.plot(unrate["DATE"].head(12),unrate["VALUE"].head(12))
plt.xticks(rotation = 90)
plt.hlines(4,"1948-01","1948-12", color = "red")
plt.xlabel("months,1948")
plt.ylabel("unemployment, rate")
plt.title("monthly  unemployment rate 1948 US ")
plt.show()
```

```python
unrate.head(12).plot("DATE","VALUE", rot = 90, title = "monthly  unemployment rate 1948 US", label = "xyz")
plt.hlines(4,"1948-01","1948-12", color = "red")
plt.xlabel("months,1948")
plt.ylabel("unemployment, rate")
```

```python
plt.plot(unrate["DATE"][12:24],unrate["VALUE"][12:24])
plt.xticks(rotation = 45)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.hlines(6.5,"1949-02","1949-12", color = "yellow")
plt.show()
unrate.head(12).plot("DATE","VALUE", rot = 90, title = "monthly  unemployment rate 1948 US", label = "xyz")
plt.hlines(4,"1948-01","1948-12", color = "red")
plt.xlabel("months,1948")
plt.ylabel("unemployment, rate")
```
