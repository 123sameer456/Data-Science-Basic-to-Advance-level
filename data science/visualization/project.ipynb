```python
#point= 1 and 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
recent = pd.read_csv("recent-grads.csv")
print(recent.info())
print(recent.head())
```

```python
#point= 3
recent.head(5)
```

```python
#point = 4
#remove all the rows where data is missing
#recent.isna()
recent.dropna()
```

```python
#point= 5
#generate a scatter plot between samplesize and median using pand
#hint : df.plot(x="",y="",kind="scatter",title="",figsize=(5,10))

recent.plot(x="Median",y="Sample_size",kind= "scatter", title= "plotting b/w samplesize and median" ,figsize= (5,7),c= "red")
plt.show()
```

```python
#point= 6
#generate a histogram of samplesize  usi g pandas.
#hint: series.plot(kind="hist", rot=40)

a=recent["Sample_size"].plot(kind="hist", rot= "40", figsize=(10,5), title="histo plotting")
a.legend()
```

```python
#generate a histo of sample size ,median, employeed and full time using for loops using pandas graphs
#all items are together
lst = ["Sample_size","Median","Employed","Full_time_year_round"]
fig = plt.figure(figsize=(5,8))
for i in range(4):
    
    a=recent[lst[i]].plot(kind="hist", rot= "40", figsize=(10,5), title="histo plotting")
    a.legend()
```

```python
#point= 7
#generate a histo of sample size ,median, employeed and full time using for loops using pandas graphs
lst = ["Sample_size","Median","Employed","Full_time_year_round"]
fig = plt.figure(figsize = (10,15))
for i in range(4):
    x = fig.add_subplot(4,1,i+1)
    x.hist(recent[lst[i]])
    x.set_title(lst[i])
```

```python
#point= 8
#generate bar diagram using first ten rows of "sharewomen". x-axis sholud mention "major" column subjects (using pandas).
#hint:  DF.head(10).plot.bar(x="",y="")
fig = plt.figure()
a=recent.head(10).plot.bar(x="Major_category",y="ShareWomen")
a.set_title("BAR GRAPH")
a.set_xlabel("Major_category")
a.set_ylabel("ShareWomen")

```
