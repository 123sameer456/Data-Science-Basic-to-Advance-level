```python
import pandas as pd
```

```python
import sqlite3 as sql
```

```python
DB = "jobs.db"
```

```python
conn = sql.connect(DB)
```

```python
q = """Select * from sqlite_master where type = 'table'"""
pd.read_sql_query(q,conn)
```

```python
def run_query(q):
    return pd.read_sql_query(q,conn)
```

```python
q = """Select * from recent_grads limit 5"""
run_query(q)
```

```python
q = """Select * from recent_grads where ShareWomen < 0.5 limit 5;"""
run_query(q)
```

```python
q = """Select * from recent_grads limit 5"""
run_query(q).columns
```

```python
q = """Select Rank,Major_code,Major_category from recent_grads where ShareWomen < 0.5 and Median > 10000 limit 5;"""
run_query(q)
```

```python
q = """Select Rank,Men,Women from recent_grads where   Median > 10000 limit 5;"""
run_query(q)
```

```python
q = """Select Rank,Major_code,Major_category,Major,Median from recent_grads where Major_category = "Engineering" or Median > 100000 limit 5;"""
run_query(q)
```

```python
q = """Select Rank,Major_code,Major_category,Major,Median from recent_grads where Major_category = "Engineering" and (ShareWomen > 0.5 or Unemployment_rate < .05) limit 5;"""
run_query(q)
```

```python
q = """Select Rank,Major_code,Major_category,Major,Median from recent_grads 
where Major_category = "Engineering" and (ShareWomen > 0.5 or Unemployment_rate < .05)
order By Median ASC
limit 5;"""
run_query(q)
```
